```python
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://csc.edu.vn/")
```

```python
#Lấy nội dung cả trang
driver.page_source
```

```python
#Lấy tiêu đề của trang
driver.title
```

```python
#Lấy đường dẫn diện tại
driver.current_url
```

TRUY XUẤT ĐỐI TƯỢNG

```python
try:
    tieu_de = driver.find_element(by=By.CLASS_NAME, value="tieu-de-muc")
except Exception as e:
    print(str(e))
else:
    print(tieu_de.text)
```

```python
try:
    tieu_de = driver.find_element(by=By.CSS_SELECTOR, value="h2.tieu-de-muc")
except Exception as e:
    print(str(e))
else:
    print(tieu_de.text)
```

```python
try:
    link_tra_cu_ket_qua_thi = driver.find_element(by=By.LINK_TEXT, value="Tra cứu kết quả thi")
except Exception as e:
    print(str(e))
else:
    print(link_tra_cu_ket_qua_thi.text)
    print(link_tra_cu_ket_qua_thi.get_attribute("href"))
```

```python
try:
    link_tra_cu_ket_qua_thi = driver.find_element(by=By.PARTIAL_LINK_TEXT, value="quả thi")
except Exception as e:
    print(str(e))
else:
    print(link_tra_cu_ket_qua_thi.text)
    print(link_tra_cu_ket_qua_thi.get_attribute("href"))
```
