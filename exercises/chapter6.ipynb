Ex 7

```python
df = pd.read_csv('../data/employees.csv')
df['HIRE_DATE'] = pd.to_datetime(df['HIRE_DATE'])
start_date = pd.to_datetime('2002-08-01')
end_date = pd.to_datetime('2002-08-31')
df_filter = df[(df['HIRE_DATE'] >= start_date) & (df['HIRE_DATE'] <= end_date)]
df_result = df_filter.sort_values(by='HIRE_DATE', ascending=False)
df_result
```
